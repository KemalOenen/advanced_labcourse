```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scienceplots
plt.style.use(['science', 'notebook', 'grid', 'high-contrast'])
import uncertainties
```

## Aufgabe 2

```python
def linear_fit(x, a, b):
    return a*x+b
#only a test but the data is right

ppb_acetonitrile=np.array([1.98, 3.96, 4.95, 5.94, 7.92])
cps_acetonitrile=np.array([10.4634, 18.4677, 23.6138, 28.5725, 37.1830])
cps_acetonitrile_uncert=np.sqrt(cps_acetonitrile)
plt.errorbar(ppb_acetonitrile, cps_acetonitrile, yerr=cps_acetonitrile_uncert, ls='')

popt, pcov=curve_fit(linear_fit, ppb_acetonitrile, cps_acetonitrile, p0=[1, 1])
plt.plot(np.arange(1.9, 8, 0.1), linear_fit(np.arange(1.9, 8, 0.1), popt[0], popt[1]))
```

```python
#calculations
percentages=np.array([0.2, 0.4, 0.5, 0.6, 0.8])
ppb_acetaldehyde=0.9*100*percentages
ppb_acetone=0.95*10*percentages
ppb_acetic_acid=1.03*10*percentages
ppb_dimethyl_sulfide=0.96*10*percentages
ppb_methyl_vinyl_ketone=0.98*10*percentages
ppb_benzene=0.98*10*percentages
ppb_3_hexanone=1.01*10*percentages
ppb_octanone=1*10*percentages
ppb_limenone=0.98*10*percentages
```

```python
cps_acetaldehyde=np.array([6.65375, 7.34019, 7.65186, 7.94390, 8.51329])*10 #done without background substraction since otherwise invisible
cps_acetone=np.array([0.827166, 1.5255, 1.91251, 2.24812, 3.01786])*10
cps_acetic_acid=np.array([0, 0, 0, 0, 0]) #nothing measured
cps_dimethyl_sulfide=np.array([0.343284, 0.862720, 1.11615, 1.42955, 1.93067])*10 #2
cps_methyl_vinyl_ketone=np.array([0.645789, 1.49431, 1.94070, 2.27303, 3.02509])*10 #3
cps_benzene=np.array([0.437319, 0.867282, 1.13925, 1.33494, 1.78869])*10 #3
cps_3_hexanone=np.array([0.496658, 1.02163, 1.32382, 1.57862, 2.10810])*10 #3
cps_octanone=np.array([0.31, 0.923459, 1.25818, 1.48314, 2.03004])*10 #2
cps_limenone=np.array([0.199718, 0.386521, 0.496227, 0.622716, 0.802845])*10 #2
```

```python
fig, axs=plt.subplots(1, 3, figsize=(20, 7))

#acetonitrile and acetone #and acetaldehyde
axs[0].errorbar(ppb_acetonitrile, cps_acetonitrile, yerr=np.sqrt(cps_acetonitrile), ls='', elinewidth=2, capsize=4, label="Acetonitril", color="blue")
popt_acetonitrile, pcov_acetonitrile=curve_fit(linear_fit, ppb_acetonitrile, cps_acetonitrile, p0=[1, 1], sigma=np.sqrt(cps_acetonitrile), absolute_sigma=True)
x_temp=np.arange(ppb_acetonitrile.min(), ppb_acetonitrile.max(), 0.1)
axs[0].plot(x_temp, linear_fit(x_temp, popt_acetonitrile[0], popt_acetonitrile[1]), color="darkblue")

axs[0].errorbar(ppb_acetone, cps_acetone, yerr=np.sqrt(cps_acetone), ls='', elinewidth=2, capsize=4, label="Aceton", color="red")
popt_acetone, pcov_acetone=curve_fit(linear_fit, ppb_acetone, cps_acetone, p0=[1, 1], sigma=np.sqrt(cps_acetone), absolute_sigma=True)
x_temp=np.arange(ppb_acetone.min(), ppb_acetone.max(), 0.1)
axs[0].plot(x_temp, linear_fit(x_temp, popt_acetone[0], popt_acetone[1]), color="darkred")

#axs[0].errorbar(ppb_acetaldehyde, cps_acetaldehyde, yerr=np.sqrt(cps_acetaldehyde), ls='', elinewidth=2, capsize=4, label="Acetaldehyd", color="green")
popt_acetaldehyde, pcov_acetaldehyde=curve_fit(linear_fit, ppb_acetaldehyde, cps_acetaldehyde, p0=[1, 1], sigma=np.sqrt(cps_acetaldehyde), absolute_sigma=True)
x_temp=np.arange(ppb_acetaldehyde.min(), ppb_acetaldehyde.max(), 0.1)
#axs[0].plot(x_temp, linear_fit(x_temp, popt_acetaldehyde[0], popt_acetaldehyde[1]), color="darkgreen")
axs[0].legend()
axs[0].set_xlabel("Konzentration / ppb")
axs[0].set_ylabel("Intensität / cps")

#_______________________________________________________________________________________________________

#methyl_vinyl_ketone and 3 hexanone and octanone
axs[1].errorbar(ppb_methyl_vinyl_ketone, cps_methyl_vinyl_ketone, yerr=np.sqrt(cps_methyl_vinyl_ketone), ls='', elinewidth=2, capsize=4, label="Methyl-Vinyl-Keton", color="blue")
popt_methyl_vinyl_ketone, pcov_methyl_vinyl_ketone=curve_fit(linear_fit, ppb_methyl_vinyl_ketone, cps_methyl_vinyl_ketone, p0=[1, 1], sigma=np.sqrt(cps_methyl_vinyl_ketone), absolute_sigma=True)
x_temp=np.arange(ppb_methyl_vinyl_ketone.min(), ppb_methyl_vinyl_ketone.max(), 0.1)
axs[1].plot(x_temp, linear_fit(x_temp, popt_methyl_vinyl_ketone[0], popt_methyl_vinyl_ketone[1]), color="darkblue")

axs[1].errorbar(ppb_3_hexanone, cps_3_hexanone, yerr=np.sqrt(cps_3_hexanone), ls='', elinewidth=2, capsize=4, label="3-hexanon", color="red")
popt_3_hexanone, pcov_3_hexanone=curve_fit(linear_fit, ppb_3_hexanone, cps_3_hexanone, p0=[1, 1], sigma=np.sqrt(cps_3_hexanone), absolute_sigma=True)
x_temp=np.arange(ppb_3_hexanone.min(), ppb_3_hexanone.max(), 0.1)
axs[1].plot(x_temp, linear_fit(x_temp, popt_3_hexanone[0], popt_3_hexanone[1]), color="darkred")

axs[1].errorbar(ppb_octanone, cps_octanone, yerr=np.sqrt(cps_octanone), ls='', elinewidth=2, capsize=4, label="Octanon", color="green")
popt_octanone, pcov_octanone=curve_fit(linear_fit, ppb_octanone, cps_octanone, p0=[1, 1], sigma=np.sqrt(cps_octanone), absolute_sigma=True)
x_temp=np.arange(ppb_octanone.min(), ppb_octanone.max(), 0.1)
axs[1].plot(x_temp, linear_fit(x_temp, popt_octanone[0], popt_octanone[1]), color="darkgreen")
axs[1].legend()
axs[1].set_xlabel("Konzentration / ppb")

#________________________________________________________________________

#benzene and dimethyl sulfide and limenone
axs[2].errorbar(ppb_benzene, cps_benzene, yerr=np.sqrt(cps_benzene), ls='', elinewidth=2, capsize=4, label="Benzen", color="blue")
popt_benzene, pcov_benzene=curve_fit(linear_fit, ppb_benzene, cps_benzene, p0=[1, 1], sigma=np.sqrt(cps_benzene), absolute_sigma=True)
x_temp=np.arange(ppb_benzene.min(), ppb_benzene.max(), 0.1)
axs[2].plot(x_temp, linear_fit(x_temp, popt_benzene[0], popt_benzene[1]), color="darkblue")

axs[2].errorbar(ppb_dimethyl_sulfide, cps_dimethyl_sulfide, yerr=np.sqrt(cps_dimethyl_sulfide), ls='', elinewidth=2, capsize=4, label="Dimethyl Sulfide", color="red")
popt_dimethyl_sulfide, pcov_dimethyl_sulfide=curve_fit(linear_fit, ppb_dimethyl_sulfide, cps_dimethyl_sulfide, p0=[1, 1], sigma=np.sqrt(cps_dimethyl_sulfide), absolute_sigma=True)
x_temp=np.arange(ppb_dimethyl_sulfide.min(), ppb_dimethyl_sulfide.max(), 0.1)
axs[2].plot(x_temp, linear_fit(x_temp, popt_dimethyl_sulfide[0], popt_dimethyl_sulfide[1]), color="darkred")

axs[2].errorbar(ppb_limenone, cps_limenone, yerr=np.sqrt(cps_limenone), ls='', elinewidth=2, capsize=4, label="Limenon", color="green")
popt_limenone, pcov_limenone=curve_fit(linear_fit, ppb_limenone, cps_limenone, p0=[1, 1], sigma=np.sqrt(cps_limenone), absolute_sigma=True)
x_temp=np.arange(ppb_limenone.min(), ppb_limenone.max(), 0.1)
axs[2].plot(x_temp, linear_fit(x_temp, popt_limenone[0], popt_limenone[1]), color="darkgreen")
axs[2].legend()

plt.show()
```

```python
FC4_std=np.array([3.905e-7, 3.92759e-07, 3.93437e-07, 3.98059e-07, 3.9512e-07])
FC4_std_total=2.99294e-07

#looks like a lot, but it could have been done with a loop

sens_acetonitrile=uncertainties.ufloat(popt_acetonitrile[0], np.sqrt(pcov_acetonitrile[0][0]))
lod_acetonitrile=2*FC4_std_total/sens_acetonitrile
print("LOD Acetonitrile = ", lod_acetonitrile, "\nSensitivity Acetonitrile = ", sens_acetonitrile)
#divide through popts in order to get the limit of detection
#acetaldehyde
sens_acetaldehyde=uncertainties.ufloat(popt_acetaldehyde[0], np.sqrt(pcov_acetaldehyde[0][0]))
lod_acetaldehyde=2*FC4_std_total/sens_acetaldehyde
print("LOD acetaldehyde = ", lod_acetaldehyde, "\nSensitivity acetaldehyde = ", sens_acetaldehyde)
#acetone
sens_acetone=uncertainties.ufloat(popt_acetone[0], np.sqrt(pcov_acetone[0][0]))
lod_acetone=2*FC4_std_total/sens_acetone
print("LOD acetone = ", lod_acetone, "\nSensitivity acetone = ", sens_acetone)
#acetic_acid is zero
#dimethyl_sulfide
sens_dimethyl_sulfide=uncertainties.ufloat(popt_dimethyl_sulfide[0], np.sqrt(pcov_dimethyl_sulfide[0][0]))
lod_dimethyl_sulfide=2*FC4_std_total/sens_dimethyl_sulfide
print("LOD dimethyl_sulfide = ", lod_dimethyl_sulfide, "\nSensitivity dimethyl_sulfide = ", sens_dimethyl_sulfide)
#methyl_vinyl_ketone
sens_methyl_vinyl_ketone=uncertainties.ufloat(popt_methyl_vinyl_ketone[0], np.sqrt(pcov_methyl_vinyl_ketone[0][0]))
lod_methyl_vinyl_ketone=2*FC4_std_total/sens_methyl_vinyl_ketone
print("LOD methyl_vinyl_ketone = ", lod_methyl_vinyl_ketone, "\nSensitivity methyl_vinyl_ketone = ", sens_methyl_vinyl_ketone)
#benzene
sens_benzene=uncertainties.ufloat(popt_benzene[0], np.sqrt(pcov_benzene[0][0]))
lod_benzene=2*FC4_std_total/sens_benzene
print("LOD benzene = ", lod_benzene, "\nSensitivity benzene = ", sens_benzene)

#3_hexanone
sens_3_hexanone=uncertainties.ufloat(popt_3_hexanone[0], np.sqrt(pcov_3_hexanone[0][0]))
lod_3_hexanone=2*FC4_std_total/sens_3_hexanone
print("LOD 3_hexanone = ", lod_3_hexanone, "\nSensitivity 3_hexanone = ", sens_3_hexanone)

#octanone
sens_octanone=uncertainties.ufloat(popt_octanone[0], np.sqrt(pcov_octanone[0][0]))
lod_octanone=2*FC4_std_total/sens_octanone
print("LOD octanone = ", lod_octanone, "\nSensitivity octanone = ", sens_octanone)

#limenone
sens_limenone=uncertainties.ufloat(popt_limenone[0], np.sqrt(pcov_limenone[0][0]))
lod_limenone=2*FC4_std_total/sens_limenone
print("LOD limenone = ", lod_limenone, "\nSensitivity limenone = ", sens_limenone)
```
