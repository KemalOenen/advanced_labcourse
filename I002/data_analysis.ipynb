```python
def linear_fit(x, a, b):
    return a*x+b
#only a test but the data is right

ppb_acetonitrile=np.array([1.98, 3.96, 4.95, 5.94, 7.92])
cps_acetonitrile=np.array([10.4634, 18.4677, 23.6138, 28.5725, 37.1830])
ppb_acetonitrile_uncert=np.sqrt(ppb_acetonitrile)
plt.errorbar(ppb_acetonitrile, cps_acetonitrile, yerr=ppb_acetonitrile_uncert, ls='')

popt, pcov=curve_fit(linear_fit, ppb_acetonitrile, cps_acetonitrile, p0=[1, 1])
plt.plot(np.arange(1.9, 8, 0.1), linear_fit(np.arange(1.9, 8, 0.1), popt[0], popt[1]))
```
