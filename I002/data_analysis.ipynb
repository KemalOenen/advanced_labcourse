```python
df=pd.read_csv(".\data_for_Guass_fits.csv")
#functions for easier fitting
def get_mass_cps(m, delta):
    array = np.asarray(df["mass"])
    index = (np.abs(array - m)).argmin()
    i1=int(index-delta)
    i2=int(index+delta)
    return df["mass"][i1:i2], df["Combined"][i1:i2]

def Gauss_fit(x, sigma, a, b):
    return (a*np.exp(-0.5*(x-b)**2 / sigma**2))

def calculate_FWHM(sigma):
    return 2*np.sqrt(2*np.log(2))*sigma

pr=230 #Pa
p0=101325 #Pa
k=uncertainties.ufloat(3, 1)*1e-9
n0=2.5e19
Flux=uncertainties.ufloat(2.92277e4, np.sqrt(2.92277e4))
def reaction_time(s):
    return s*p0/Flux/k/n0
```

```python
#this part requires redifining the functions from before
def get_mass(m, delta, df):
    array = np.asarray(df["mass"])
    index = (np.abs(array - m)).argmin()
    i1=int(index-delta)
    i2=int(index+delta)
    return df["mass"][i1:i2], df["Combined"][i1:i2]
```
